```python
all_inputs = iris_data[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']].values
```

```python
all_inputs[:10]
```

```python
all_labels = iris_data['class'].values
```

```python
all_labels[:10]
```

```python
from sklearn.model_selection import train_test_split
```

```python
(training_inputs, testing_inputs, training_labels, testing_labels) = train_test_split(all_inputs, all_labels, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier()
```

```python
decision_tree.fit(training_inputs, training_labels)
```

```python
decision_tree.score(testing_inputs, testing_labels)
```

```python
accuracies = []
for repeatition in range(1000):
    (training_inputs, testing_inputs, training_labels, testing_labels) = train_test_split(all_inputs, all_labels, test_size=0.25)
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(training_inputs, training_labels)
    score = decision_tree.score(testing_inputs, testing_labels)
    accuracies.append(score)

plt.hist(accuracies);
```
