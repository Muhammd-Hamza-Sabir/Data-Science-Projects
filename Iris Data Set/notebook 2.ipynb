<h2>Data Visualization on Iris</h2>

```python
# importing libraries
import pandas as pd
import warnings
warnings.filterwarnings(action="ignore")
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading data
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.set_index(keys='Id', drop=True, inplace=True)
```

```python
df.head()
```

```python
# get descriptice statistics of all the variables
df.describe(include='all')
```

```python
# as there is no missing values
# for more insights, let's do some EDA
# scatterplot using pandas
df.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# same can be done by using seaborn
sns.jointplot(data=df, x="SepalWidthCm", y="SepalLengthCm")
```

```python
sns.jointplot(data=df, y="PetalLengthCm", x="PetalWidthCm")
```

```python
# to make them distinguishabale, draw using facetgrid
sns.FacetGrid(data=df, hue="Species", size=6).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
```

```python
# plot shows that PetalLength and PetalWidth are better combination for classification than above two features
sns.FacetGrid(data=df, hue="Species", size=6).map(plt.scatter, "PetalWidthCm", "PetalLengthCm").add_legend()
```

```python
# boxplot and points on each boxplot
plt.figure(figsize=(10,6))
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=df)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=df, jitter=True)
```

```python
# above can be achieved by this songle plot
# graph show there is overlapping between versicolr and virginica's petal length while setosa's petal length has different range
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="Species", y="PetalLengthCm")
```

```python
# let's plot all the features at one place
sns.pairplot(data=df, hue="Species")
```

```python
# we are done with seaborn visualization
# let's try some pandas plotting methods
df.boxplot(by="Species", figsize=(12, 6))
```

```python
#Now let us see how are the length and width are distributed
df.hist(edgecolor='black', linewidth=1.2)
fig = plt.gcf()
fig.set_size_inches(12, 6)
plt.show()
```

```python
# Another multivariate visualization technique pandas has is parallel_coordinates
# Parallel coordinates plots each feature on a separate column & then draws lines
# connecting the features for each data sample
from pandas.tools.plotting import parallel_coordinates
plt.figure(figsize=(10,6))
parallel_coordinates(df, "Species")
```
